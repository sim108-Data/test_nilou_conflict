### Cardiorespiratory Fitnes

#### FU2

```python
# PAL coefficient (Raichlen et al.)

df['F2PAL'] = df['F2MVPA']*7/30 # nombres de sessions de 30mn de MVPA par semaine
def PALs(x) : 
        if x == 0 :
            return 0
        if 0< x <=1.5:
            return 0.39
        if 1.5< x <=3:
            return 0.35
        if 3< x <=6:
            return 0.64
        if x >6:
            return 1.21  
df["F2PALr"]= df.F2PAL.apply(lambda x : PALs(x))
df['F2CRF'] = 3.5 * ((2.78 * df['F2sex']) - (0.11 * df['F2age']) - (0.17 * df['F2BMI']) - (0.05 * df['F2HRTRTE']) + df['F2PALr'] + 21.41)
df['F2CRF'] = df['F2CRF'].astype('float')
df = df.drop(columns=['F2PAL','F2PALr'], axis = 1)

#equation eCRF avec multiple de 3.5 : 1 MET = 3.5ml/O2..
```

#### FU3

```python
# PAL coefficient (Raichlen et al.)

df['F3PAL'] = df['F3MVPA']*7/30 # nombres de sessions de 30mn de MVPA par semaine
def PALs(x) : 
        if x == 0 :
            return 0
        if 0< x <=1.5:
            return 0.39
        if 1.5< x <=3:
            return 0.35
        if 3< x <=6:
            return 0.327534
        if x >6:
            return 1.21  
df["F3PALr"]= df.F3PAL.apply(lambda x : PALs(x))
df['F3CRF'] = 3.5 * ((2.78 * df['F3sex']) - (0.11 * df['F3age']) - (0.17 * df['F3BMI']) - (0.05 * df['F3HRTRTE']) + df['F3PALr'] + 21.41)
df['F3CRF'] = df['F3CRF'].astype('float')
df = df.drop(columns=['F3PAL','F3PALr'], axis = 1)

#equation eCRF avec multiple de 3.5 : 1 MET = 3.5ml/O2..
```

```python
df['d_CRF'] = df['F3CRF'] - df['F2CRF']
```

```python
# Function that creates a correlation matrix with Pearson's correlation between continuous and PointBiseral between categorical and/or continous , it also returns the Bonferroni-corrected p-values

def corr_matrix(df, dich, cont):
    dfc = pd.DataFrame(columns=df.columns)
    coefs = dfc.transpose().join(dfc, how='outer')
    pvals = dfc.transpose().join(dfc, how='outer')
    for r in df.columns:
        for c in df.columns:
            x = df[r]
            y = df[c]
            bad = ~np.logical_or(np.isnan(x), np.isnan(y))
            X = np.compress(bad, x)
            Y = np.compress(bad, y)
            if (r in cont) and (c in cont):
                coefs[r][c] = pearsonr(X,Y)[0]
                pvals[r][c] = pearsonr(X,Y)[1]
            else:
                coefs[r][c] = pointbiserialr(X,Y)[0]
                pvals[r][c] = pointbiserialr(X,Y)[1]
    Bonf_pvals = pvals * sum(range(len(pvals.columns))) # Bonferroni correction - multiplying pvals by nb of coefs
    coefs = coefs.astype(float)
    pvals = pvals.astype(float)
    Bonf_pvals = Bonf_pvals.astype(float)
    return coefs, pvals, Bonf_pvals

# Create the correlation matrix

dfCorr = MY_2.filter(['F2sex', 'F2age','FU2c_MDDPD2','MDDnew','MDD2_LO','F2_GAF_L', 'F2_GAF_C', 'F2_GAF_W', 'F2MVPA','F2CRF','TIV', 'x3rdVentricle', 'x4thVentricle', 'BrainStem', 'CSF', 'AccumbensArea', 'Caudate', 'CerebellumExterior', 'CerebralWhiteMatter', 'InfLatVent', 'LateralVentricle', 'Pallidum', 'Putamen', 'BasalForebrain', 'ACgGAnteriorCingulateGyrus', 'AInsAnteriorInsula', 'AOrGAnteriorOrbitalGyrus', 'AnGAngularGyrus', 'CalcCalcarineCortex', 'COCentralOperculum', 'CunCuneus', 'EntEntorhinalArea', 'FOFrontalOperculum', 'FRPFrontalPole', 'FuGFusiformGyrus', 'RightGReGyrusRectus', 'IOGInferiorOccipitalGyrus', 'ITGInferiorTemporalGyrus', 'LiGLingualGyrus', 'LOrGLateralOrbitalGyrus', 'MCgGMiddleCingulateGyrus', 'MFCMedialFrontalCortex', 'MFGMiddleFrontalGyrus', 'MOGMiddleOccipitalGyrus', 'MOrGMedialOrbitalGyrus', 'MPoGPostcentralGyrusMedialSegment', 'MPrGPrecentralGyrusMedialSegment', 'MSFGSuperiorFrontalGyrusMedialSegment', 'MTGMiddleTemporalGyrus', 'OCPOccipitalPole', 'OFuGOccipitalFusiformGyrus', 'OpIFGOpercularPartOfTheInferiorFrontalGyrus', 'OrIFGOrbitalPartOfTheInferiorFrontalGyrus', 'PCgGPosteriorCingulateGyrus', 'PCuPrecuneus', 'PHGParahippocampalGyrus', 'PInsPosteriorInsula', 'POParietalOperculum', 'PoGPostcentralGyrus', 'POrGPosteriorOrbitalGyrus', 'PPPlanumPolare', 'PrGPrecentralGyrus', 'PTPlanumTemporale', 'SCASubcallosalArea', 'SFGSuperiorFrontalGyrus', 'SMCSupplementaryMotorCortex', 'SMGSupramarginalGyrus', 'SOGSuperiorOccipitalGyrus', 'SPLSuperiorParietalLobule', 'STGSuperiorTemporalGyrus', 'TMPTemporalPole', 'TrIFGTriangularPartOfTheInferiorFrontalGyrus', 'TTGTransverseTemporalGyrus', 'STN', 'SubstantiaNigra', 'RedNucleus', 'DentateNucleus', 'Th_anterior', 'Th_geniculate', 'Th_mediodorsal', 'Th_intralaminar', 'Th_pulvinar', 'Th_ventroanterior', 'Th_ventrolateral', 'Th_ventroposterior', 'Th_posteriorAnterior', 'Agd_laterobasal', 'Agd_superficial', 'Hi_SPAM_CA10x2D3', 'Hi_SPAM_CA40x2DDG', 'Hi_SPAM_subiculum'])
dich =['F2sex','FU2c_MDDPD2','MDDnew','MDD2_LO']
cont = ['F2age','F2_GAF_L', 'F2_GAF_C', 'F2_GAF_W','F2MVPA','F2CRF','TIV', 'x3rdVentricle', 'x4thVentricle', 'BrainStem', 'CSF', 'AccumbensArea', 'Caudate', 'CerebellumExterior', 'CerebralWhiteMatter', 'InfLatVent', 'LateralVentricle', 'Pallidum', 'Putamen', 'BasalForebrain', 'ACgGAnteriorCingulateGyrus', 'AInsAnteriorInsula', 'AOrGAnteriorOrbitalGyrus', 'AnGAngularGyrus', 'CalcCalcarineCortex', 'COCentralOperculum', 'CunCuneus', 'EntEntorhinalArea', 'FOFrontalOperculum', 'FRPFrontalPole', 'FuGFusiformGyrus', 'RightGReGyrusRectus', 'IOGInferiorOccipitalGyrus', 'ITGInferiorTemporalGyrus', 'LiGLingualGyrus', 'LOrGLateralOrbitalGyrus', 'MCgGMiddleCingulateGyrus', 'MFCMedialFrontalCortex', 'MFGMiddleFrontalGyrus', 'MOGMiddleOccipitalGyrus', 'MOrGMedialOrbitalGyrus', 'MPoGPostcentralGyrusMedialSegment', 'MPrGPrecentralGyrusMedialSegment', 'MSFGSuperiorFrontalGyrusMedialSegment', 'MTGMiddleTemporalGyrus', 'OCPOccipitalPole', 'OFuGOccipitalFusiformGyrus', 'OpIFGOpercularPartOfTheInferiorFrontalGyrus', 'OrIFGOrbitalPartOfTheInferiorFrontalGyrus', 'PCgGPosteriorCingulateGyrus', 'PCuPrecuneus', 'PHGParahippocampalGyrus', 'PInsPosteriorInsula', 'POParietalOperculum', 'PoGPostcentralGyrus', 'POrGPosteriorOrbitalGyrus', 'PPPlanumPolare', 'PrGPrecentralGyrus', 'PTPlanumTemporale', 'SCASubcallosalArea', 'SFGSuperiorFrontalGyrus', 'SMCSupplementaryMotorCortex', 'SMGSupramarginalGyrus', 'SOGSuperiorOccipitalGyrus', 'SPLSuperiorParietalLobule', 'STGSuperiorTemporalGyrus', 'TMPTemporalPole', 'TrIFGTriangularPartOfTheInferiorFrontalGyrus', 'TTGTransverseTemporalGyrus', 'STN', 'SubstantiaNigra', 'RedNucleus', 'DentateNucleus', 'Th_anterior', 'Th_geniculate', 'Th_mediodorsal', 'Th_intralaminar', 'Th_pulvinar', 'Th_ventroanterior', 'Th_ventrolateral', 'Th_ventroposterior', 'Th_posteriorAnterior', 'Agd_laterobasal', 'Agd_superficial', 'Hi_SPAM_CA10x2D3', 'Hi_SPAM_CA40x2DDG', 'Hi_SPAM_subiculum']

coefs, pvals, Bonf_pvals = corr_matrix(dfCorr, dich, cont)

# Plot the heatmap annotated with correlation coefficients, in bold when Bonferroni-significant
xticks=['F2sex', 'F2age','FU2c_MDDPD2','MDDnew','MDD2_LO','F2_GAF_L', 'F2_GAF_C', 'F2_GAF_W', 'F2MVPA','F2CRF','TIV', 'x3rdVentricle', 'x4thVentricle', 'BrainStem', 'CSF', 'AccumbensArea', 'Caudate', 'CerebellumExterior', 'CerebralWhiteMatter', 'InfLatVent', 'LateralVentricle', 'Pallidum', 'Putamen', 'BasalForebrain', 'ACgGAnteriorCingulateGyrus', 'AInsAnteriorInsula', 'AOrGAnteriorOrbitalGyrus', 'AnGAngularGyrus', 'CalcCalcarineCortex', 'COCentralOperculum', 'CunCuneus', 'EntEntorhinalArea', 'FOFrontalOperculum', 'FRPFrontalPole', 'FuGFusiformGyrus', 'RightGReGyrusRectus', 'IOGInferiorOccipitalGyrus', 'ITGInferiorTemporalGyrus', 'LiGLingualGyrus', 'LOrGLateralOrbitalGyrus', 'MCgGMiddleCingulateGyrus', 'MFCMedialFrontalCortex', 'MFGMiddleFrontalGyrus', 'MOGMiddleOccipitalGyrus', 'MOrGMedialOrbitalGyrus', 'MPoGPostcentralGyrusMedialSegment', 'MPrGPrecentralGyrusMedialSegment', 'MSFGSuperiorFrontalGyrusMedialSegment', 'MTGMiddleTemporalGyrus', 'OCPOccipitalPole', 'OFuGOccipitalFusiformGyrus', 'OpIFGOpercularPartOfTheInferiorFrontalGyrus', 'OrIFGOrbitalPartOfTheInferiorFrontalGyrus', 'PCgGPosteriorCingulateGyrus', 'PCuPrecuneus', 'PHGParahippocampalGyrus', 'PInsPosteriorInsula', 'POParietalOperculum', 'PoGPostcentralGyrus', 'POrGPosteriorOrbitalGyrus', 'PPPlanumPolare', 'PrGPrecentralGyrus', 'PTPlanumTemporale', 'SCASubcallosalArea', 'SFGSuperiorFrontalGyrus', 'SMCSupplementaryMotorCortex', 'SMGSupramarginalGyrus', 'SOGSuperiorOccipitalGyrus', 'SPLSuperiorParietalLobule', 'STGSuperiorTemporalGyrus', 'TMPTemporalPole', 'TrIFGTriangularPartOfTheInferiorFrontalGyrus', 'TTGTransverseTemporalGyrus', 'STN', 'SubstantiaNigra', 'RedNucleus', 'DentateNucleus', 'Th_anterior', 'Th_geniculate', 'Th_mediodorsal', 'Th_intralaminar', 'Th_pulvinar', 'Th_ventroanterior', 'Th_ventrolateral', 'Th_ventroposterior', 'Th_posteriorAnterior', 'Agd_laterobasal', 'Agd_superficial', 'Hi_SPAM_CA10x2D3', 'Hi_SPAM_CA40x2DDG', 'Hi_SPAM_subiculum']
yticks=['F2sex', 'F2age','FU2c_MDDPD2','MDDnew','MDD2_LO','F2_GAF_L', 'F2_GAF_C', 'F2_GAF_W', 'F2MVPA','F2CRF','TIV', 'x3rdVentricle', 'x4thVentricle', 'BrainStem', 'CSF', 'AccumbensArea', 'Caudate', 'CerebellumExterior', 'CerebralWhiteMatter', 'InfLatVent', 'LateralVentricle', 'Pallidum', 'Putamen', 'BasalForebrain', 'ACgGAnteriorCingulateGyrus', 'AInsAnteriorInsula', 'AOrGAnteriorOrbitalGyrus', 'AnGAngularGyrus', 'CalcCalcarineCortex', 'COCentralOperculum', 'CunCuneus', 'EntEntorhinalArea', 'FOFrontalOperculum', 'FRPFrontalPole', 'FuGFusiformGyrus', 'RightGReGyrusRectus', 'IOGInferiorOccipitalGyrus', 'ITGInferiorTemporalGyrus', 'LiGLingualGyrus', 'LOrGLateralOrbitalGyrus', 'MCgGMiddleCingulateGyrus', 'MFCMedialFrontalCortex', 'MFGMiddleFrontalGyrus', 'MOGMiddleOccipitalGyrus', 'MOrGMedialOrbitalGyrus', 'MPoGPostcentralGyrusMedialSegment', 'MPrGPrecentralGyrusMedialSegment', 'MSFGSuperiorFrontalGyrusMedialSegment', 'MTGMiddleTemporalGyrus', 'OCPOccipitalPole', 'OFuGOccipitalFusiformGyrus', 'OpIFGOpercularPartOfTheInferiorFrontalGyrus', 'OrIFGOrbitalPartOfTheInferiorFrontalGyrus', 'PCgGPosteriorCingulateGyrus', 'PCuPrecuneus', 'PHGParahippocampalGyrus', 'PInsPosteriorInsula', 'POParietalOperculum', 'PoGPostcentralGyrus', 'POrGPosteriorOrbitalGyrus', 'PPPlanumPolare', 'PrGPrecentralGyrus', 'PTPlanumTemporale', 'SCASubcallosalArea', 'SFGSuperiorFrontalGyrus', 'SMCSupplementaryMotorCortex', 'SMGSupramarginalGyrus', 'SOGSuperiorOccipitalGyrus', 'SPLSuperiorParietalLobule', 'STGSuperiorTemporalGyrus', 'TMPTemporalPole', 'TrIFGTriangularPartOfTheInferiorFrontalGyrus', 'TTGTransverseTemporalGyrus', 'STN', 'SubstantiaNigra', 'RedNucleus', 'DentateNucleus', 'Th_anterior', 'Th_geniculate', 'Th_mediodorsal', 'Th_intralaminar', 'Th_pulvinar', 'Th_ventroanterior', 'Th_ventrolateral', 'Th_ventroposterior', 'Th_posteriorAnterior', 'Agd_laterobasal', 'Agd_superficial', 'Hi_SPAM_CA10x2D3', 'Hi_SPAM_CA40x2DDG', 'Hi_SPAM_subiculum']

fig, ax = plt.subplots(figsize=(300,200))
plt.title("FU2 correlation between PA, MDD and Brain tissues", fontsize= 40, pad=40)
sns.heatmap(coefs, mask=abs(Bonf_pvals)<0.05, annot=True, annot_kws={'size': 15}, fmt=".2f", ax=ax,xticklabels=xticks,yticklabels=yticks, cmap='PiYG', vmin=-1, vmax=1)
sns.heatmap(coefs, mask=abs(Bonf_pvals)>=0.05, cbar=False, annot=True, annot_kws={"weight": "bold", 'size': 15},fmt=".2f", ax=ax,xticklabels=xticks, yticklabels=yticks, cmap='PiYG', vmin=-1, vmax=1)
plt.xticks(rotation=45, ha='right', fontsize=18)
plt.yticks(fontsize=18)
ax.figure.axes[-1].yaxis.label.set_size(22)
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=18)
plt.show()



# Save the figure
fig.savefig('FU2 PA, MDD, Myelin correlation.png')
```
