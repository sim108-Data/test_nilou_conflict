```python
# Moderate-to-vigorous physical activity
df4 = pd.read_csv('/Users/Nilavan/Documents/Jupyter/MD_share/Data/PA_2.csv', sep=';').filter(['F2pt', 'F2dur_day', 'F2Weekend', 'F2dur_toinday','F2dur_tmodday', 'F2dur_tvigday'])
df4['valid'] = df4['F2dur_day']>=600
df4['F2_MVPA'] = df4['F2dur_tmodday'] + df4['F2dur_tvigday']

# Valid day = minimum 600 min; valid subject = minimum 5 valid week days and 2 valid weekend days
PA2 = pd.DataFrame(index=range(len(df4['F2pt'].unique())), columns=['pt', 'F2MVPA'])
PA2['pt'] = df4['F2pt'].unique()

for pt in df4['F2pt'].unique():
    if (len(df4.loc[(df4['F2pt']==pt) & (df4['F2Weekend']==0) & (df4['valid'])]) >= 5) & (len(df4.loc[(df4['F2pt']==pt) & (df4['F2Weekend']==1) & (df4['valid'])]) >= 2):
        PA2.loc[PA2['pt']==pt, 'F2MVPA'] = df4.loc[(df4['F2pt']==pt) & (df4['valid']) & (df4['F2_MVPA'].notna()), 'F2_MVPA'].mean()

PA2['F2MVPA'] = pd.to_numeric(PA2['F2MVPA'], errors='coerce')
PA2 = PA2.dropna()

PA2 = PA2.rename(columns = {'pt': 'FAM'})
PA2['FAM'] = PA2['FAM'] + 50000
PA2
```

### Cardiorespiratory Fitnes

#### FU2

```python
# PAL coefficient (Raichlen et al.)

df['F2PAL'] = df['F2MVPA']*7/30 # nombres de sessions de 30mn de MVPA par semaine
def PALs(x) : 
        if x == 0 :
            return 0
        if 0< x <=1.5:
            return 0.39
        if 1.5< x <=3:
            return 0.35
        if 3< x <=6:
            return 0.64
        if x >6:
            return 1.21  
df["F2PALr"]= df.F2PAL.apply(lambda x : PALs(x))
df['F2CRF'] = 3.5 * ((2.78 * df['F2sex']) - (0.11 * df['F2age']) - (0.17 * df['F2BMI']) - (0.05 * df['F2HRTRTE']) + df['F2PALr'] + 21.41)
df['F2CRF'] = df['F2CRF'].astype('float')
df = df.drop(columns=['F2PAL','F2PALr'], axis = 1)

#equation eCRF avec multiple de 3.5 : 1 MET = 3.5ml/O2..
```

#### FU3

```python
# PAL coefficient (Raichlen et al.)

df['F3PAL'] = df['F3MVPA']*7/30 # nombres de sessions de 30mn de MVPA par semaine
def PALs(x) : 
        if x == 0 :
            return 0
        if 0< x <=1.5:
            return 0.39
        if 1.5< x <=3:
            return 0.35
        if 3< x <=6:
            return 0.64
        if x >6:
            return 1.21  
df["F3PALr"]= df.F3PAL.apply(lambda x : PALs(x))
df['F3CRF'] = 3.5 * ((2.78 * df['F3sex']) - (0.11 * df['F3age']) - (0.17 * df['F3BMI']) - (0.05 * df['F3HRTRTE']) + df['F3PALr'] + 21.41)
df['F3CRF'] = df['F3CRF'].astype('float')
df = df.drop(columns=['F3PAL','F3PALr'], axis = 1)

#equation eCRF avec multiple de 3.5 : 1 MET = 3.5ml/O2..
```

```python
df['d_CRF'] = df['F3CRF'] - df['F2CRF']
```

```python
import pandas as pd
import scipy.stats as stats

def compare_columns(df1, df2):
    results = []
    for col1, col2 in zip(df1.columns, df2.columns):
        if (df1[col1].nunique() == 2 ) and (df2[col2].nunique() == 2):
            chi2, p, dof, ex = stats.chi2_contingency(pd.crosstab(df1[col1], df2[col2]))
            results.append((col1, col2, 'Chi2', chi2, p))
        else:
            t, p = stats.ttest_ind(df1[col1], df2[col2])
            results.append((col1, col2, 'T-test', t, p))
    return pd.DataFrame(results, columns=['Non-BrainLaus', 'BrainLaus', 'test', 'value', 'p-value'])

df1 = df_NGM.filter(['F2age', 'F3age','SC_risk_5yrs_F3','F2BMI','F3BMI', 'FU2c_MDDPD2', 'FU2c_MDDPDon', 'MDD2_EO', 'MDD2_LO', 'MDD2_EOM', 'MDD2_EOA','MDD2_EOU', 'MDD2_LOM', 'MDD2_LOA', 'MDD2_LOU', 'FU3c_MDDPD2','FU3c_MDDPDon', 'MDD3_atyp', 'MDD3_mel', 'MDD3_unsp', 'MDD3_EO','MDD3_LO', 'MDDnew','F2_GAF_L', 'F2_GAF_C', 'F2_GAF_W', 'F3_GAF_L', 'F3_GAF_C','F3_GAF_W', 'GAF_C', 'GAF_L', 'GAF_W', 'F2MVPA', 'F3MVPA', 'MVPA','F2CRF', 'F3CRF', 'CRF'])
df2 = df_GM.filter(['F2age', 'F3age','SC_risk_5yrs_F3','F2BMI','F3BMI', 'FU2c_MDDPD2', 'FU2c_MDDPDon', 'MDD2_EO', 'MDD2_LO', 'MDD2_EOM', 'MDD2_EOA','MDD2_EOU', 'MDD2_LOM', 'MDD2_LOA', 'MDD2_LOU', 'FU3c_MDDPD2','FU3c_MDDPDon', 'MDD3_atyp', 'MDD3_mel', 'MDD3_unsp', 'MDD3_EO','MDD3_LO', 'MDDnew','F2_GAF_L', 'F2_GAF_C', 'F2_GAF_W', 'F3_GAF_L', 'F3_GAF_C','F3_GAF_W', 'GAF_C', 'GAF_L', 'GAF_W', 'F2MVPA', 'F3MVPA', 'MVPA','F2CRF', 'F3CRF', 'CRF'])

result_df = compare_columns(df1, df2)
#result_df.to_csv('Distribution BL.csv')
print(result_df)

```

### Select features, define X and y response variable

```python
y
```

```python
# Nilavan: Fit Lasso on many models, compare AIC
# Refit on the best model
Y=y
X=X.fillna(X.median()) # y is endogenous to our model, x is exogenous
y=y.fillna(y.median())

scaler = preprocessing.MinMaxScaler()

# Interaction table with polynomials
interaction = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)
X_inter=pd.DataFrame(interaction.fit_transform(X), columns=interaction.get_feature_names(input_features=X.columns)) # Table

X_inter_raw=X_inter
Y_raw=Y

# AIC
lasso_lars_ic = make_pipeline(scaler, LassoLarsIC(criterion="aic", max_iter=1000, normalize=False)).fit(X_inter_raw, Y_raw)
selected_features=X_inter.columns[lasso_lars_ic[-1].coef_!=0]

# Scaling
d = scaler.fit_transform(X_inter)
scaled_df = pd.DataFrame(d, columns= X_inter.columns)
scaled_df.index=X.index
X_inter=scaled_df
Y=pd.DataFrame(scaler.fit_transform(Y.values.reshape(-1, 1)), columns = [y.name]) # numpy environment complication
X_inter.index=Y.index

X_new=sm.add_constant(X_inter[selected_features])

# Fit final regression model with stats output
X_new.index=Y.index
model=sm.OLS(Y, X_new).fit()
print(model.summary())
```

```python
model_base=sm.OLS(Y, sm.add_constant(X_inter[x_variables+ ['F3_age d_CRF']])).fit() # only main effects of the base variables
print(model_base.summary())
```

```python
# Feature importance, B coefficient effect sizes, only significant ones
p_v_thresh=0.05

feature_imp=model.params[model.pvalues<p_v_thresh]
feature_imp=feature_imp[feature_imp.index!='const'] # drop constant term
feature_imp=abs(feature_imp).sort_values(ascending=False)

%matplotlib inline
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('effect sizes')
plt.ylabel('Features')
plt.title("Difference contribution to CRF of each risk factors (stat. signif.)")
#plt.legend()
plt.show()
```

```python
# Nilavan: Fit Lasso on many models, compare AIC
# Refit on the best model
Y=y
X=X.fillna(X.median()) # y is endogenous to our model, x is exogenous
y=y.fillna(y.median())

scaler = preprocessing.MinMaxScaler()

# Interaction table with polynomials
interaction = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)
X_inter=pd.DataFrame(interaction.fit_transform(X), columns=interaction.get_feature_names(input_features=X.columns)) # Table

X_inter_raw=X_inter
Y_raw=Y

# AIC
lasso_lars_ic = make_pipeline(scaler, LassoLarsIC(criterion="aic", max_iter=1000, normalize=False)).fit(X_inter_raw, Y_raw)
selected_features=X_inter.columns[lasso_lars_ic[-1].coef_!=0]

# Scaling
d = scaler.fit_transform(X_inter)
scaled_df = pd.DataFrame(d, columns= X_inter.columns)
scaled_df.index=X.index
X_inter=scaled_df
Y=pd.DataFrame(scaler.fit_transform(Y.values.reshape(-1, 1)), columns = [y.name]) # numpy environment complication
X_inter.index=Y.index

X_new=sm.add_constant(X_inter[selected_features])

# Fit final regression model with stats output
X_new.index=Y.index
model=sm.OLS(Y, X_new).fit()
print(model.summary())
```

```python
model_base=sm.OLS(Y, sm.add_constant(X_inter[x_variables+ ['F3_age d_MVPA']])).fit() # only main effects of the base variables
print(model_base.summary())
```

```python
# Feature importance, B coefficient effect sizes, only significant ones
p_v_thresh=0.05

feature_imp=model.params[model.pvalues<p_v_thresh]
feature_imp=feature_imp[feature_imp.index!='const'] # drop constant term
feature_imp=abs(feature_imp).sort_values(ascending=False)

%matplotlib inline
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('effect sizes')
plt.ylabel('Features')
plt.title("Difference contribution to MVPA of each risk factors (stat. signif.)")
#plt.legend()
plt.show()
```
